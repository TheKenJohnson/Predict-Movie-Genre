The first step for this project is loading in the data set. I read the file in using csv.DictReader similar to homework 2. I next filtered out any points that contained unknown genres or genres outside of the 4 labels I classify. I also decided to limit the data to only the genre and plot fields as I will not be using the other fields such as title, wikipedia entry, actors, ethnicity, etc. I now have a sample size of 3500 training entries which is ready for NLP. This contains 900 of all 4 genres except romance which has 800 due to slightly less data.

Next, I loaded in a file of stopwords. This contains 733 words that do not provide much value to NLP and will not be included in the bag of words. Most of these are words like: (the, I, us, are) which will likely be present in all genres so I dont want to account for them.

Finally, I load in a file of common names. The file contains 1000 common names I want to filter out of NLP as a name being present should not be useful to classify a movie genre.

```python
import csv
#relevant_genres = ['western', 'comedy', 'drama', 'adventure', 'horror', 'crime', 'romance', 'thriller', 'action'] 
relevant_genres = ['romance', 'comedy', 'horror', 'action']

def load_and_filter(filename):
    train = []
    valid = []
    genres = {'romance':100, 'comedy':0, 'action':0, 'horror':0}
    with open(filename, newline='', encoding = 'utf-8') as file:
        reader = csv.DictReader(file)
        for line in reader:
            if not line['Genre'] in relevant_genres or genres[line['Genre']] >= 1000 :
                continue
            if genres[line['Genre']] >= 900:
                valid.append({'Genre' : line['Genre'], 'Plot' : line['Plot']})
            else:
                train.append({'Genre' : line['Genre'], 'Plot' : line['Plot']})
            genres[line['Genre']] += 1
    return train, valid

def load_stopwords(filename):
    stopwords = []
    words = open(filename)
    for w in words:
        stopwords.append(w.replace('\n', ''))
    return stopwords

def load_names(filename):
    names = []
    with open(filename, newline='', encoding = 'utf-8') as file:
        reader = csv.DictReader(file)
        for line in reader:
            names.append(line['Name'].lower())
    return names

def genre_counts(data):
    counts = {}
    for d in data:
        if d['Genre'] not in counts.keys():
            counts[d['Genre']] = 1
        else:
            counts[d["Genre"]] += 1
    return counts

#loading and printing raw data
training, valid = load_and_filter('wiki_movie_plots_deduped.csv')
print("Training length:", len(training))
print("First training example:", training[0], "\n")

#loading and printing stopwords
stopwords = load_stopwords('stopwords.txt')
print("Stopwords:", stopwords)
print ("Length of stopwords:", len(stopwords), "\n")

#loading and printing names
names = load_names('names.csv')
print("Names:", names)
print("Length of Names:", len(names), "\n")

#printing genre counts
print("Genre counts of training:", genre_counts(training))
print("Genre counts of valid:", genre_counts(valid))
```

Next, I pass in the filtered data (which is simply a dictionary with a genre and a plot as shown above), and perform natural language processing using a bag of words approach. I start by removing all punctuation from each synopsis as that was causing issues with word frequency. After this, I split the synopsis into a set of words and make all words lowercase. From here, I can easily increment the words dictionary (which contains the word and its document frequency) if a word exists in an entry and is not present in either the list of stopwords or the list of names. Keep in mind I decided to use document frequency and not overall frequency meaning that if a word is present in a synopsis multiple times it will still only be incremented by one. 

The final step is now using the words dictionary to create the bag of words. I want to add words that are present in enough documents to be useful for classification, but not present in too many to become non impactful. After some testing and paramater tuning, I found including words with > 40 document frequency and < 1000 document frequency to produce the best results. 

```python
import string
def generate_bag_of_words(data, lower, upper):
    translator = str.maketrans('', '', string.punctuation)
    words = {}
    for d in range(len(data)):
        plot = []
        for word in data[d]['Plot'].split():
            if word not in plot:
                plot.append(word)
        for w in plot:
            word = (w).translate(translator).lower()
            if word not in stopwords and word not in names:
                if not word in words.keys():
                    words[word] = 1
                else:
                    words[word] += 1
    file = open("wordbag.txt", "w", encoding="utf-8")
    features = []
    for word in words:
        if words[word] > lower and words[word] < upper:
            print(word, words[word], file=file)
            features.append(word)
    return features

features = generate_bag_of_words(training, 45, 1000)
print("Bag of Words:", features)
print("Bag of Words length", len(features))
```

Now that the bag of words has been generated, I need to use it to convert each point from a dictionary of genre and synopsis, to a dictionary of genre, and features (where each feature will be a word in the bag of words and 0/1 based on if it is present in the synopsis). To do this, I once again split the plot into a set of lowercase words without punctuation. I then loop through the bag of words, and if each word exists in the set, it is set to 1. Finally I remove, the plot entry from the dictionary as it is no longer needed. 

```python
def addFeatures(data, features):
    translator = str.maketrans('', '', string.punctuation)
    for d in data:
        plot = set(d['Plot'].translate(translator).lower().split())
        for feature in features:
            if feature in plot:
                d[feature] = 1
            else:
                d[feature] = 0
        d.pop('Plot')
    return data

training = addFeatures(training, features)
valid = addFeatures(valid, features)
print("New format of training data:", training[0])
```

Similar to HW#2, I convert the several thousand features from the bag of words into one compact list of weights corresponding to each feature. The initial 1.0 is the bias term, and the rest of the features appended after that. As you can see below, now each training example has the format {'label': label#, 'features: [bias, feature#1, feature#2, ...feature#n]}. This format can now be used in softmax logistic regression.

```python
def extract_features(raw):
    data = []
    for r in raw:
        point = {}
        point["label"] = relevant_genres.index(r['Genre'])
        features = []
        features.append(1.)
        for feat in r.keys():
            count = 0
            if feat == 'Genre':
                continue
            features.append(r[feat])
        point['features'] = features
        data.append(point)
    return data

training = extract_features(training)
valid = extract_features(valid)
print("New format of training example", training[0])
```

The functions below are used to train the model itself. I will describe each in more detail below.

```python
import math
import random
import matplotlib.pyplot as plot


#Creates one set of weights for each genre in the list of genres. Each set of weights is length k which corresponds to the length of the features.
def initialize_models(k, genres):
    models = []
    for i in range(genres):
        models.append([random.gauss(0, 1) for x in range(k)])
    return models


#Simple dot product function needed for logistic regression. This is the same as the dot product function I implemented in HW#2
def dot(x, y):
    s = 0
    for i in range(len(x)):
        s+= x[i] * y[i]
    return s


#Performs a softmax prediction. Calculates the logistic for each model, and returns a list containing the probability of each label.
def predict(model, point):
    dot_products = []
    for c in range(len(relevant_genres)):
        dot_products.append(dot(model[c], point['features']))


    exponentials = []
    for d in dot_products:
        exponentials.append(math.exp(d))


    probabilities = []
    total = sum(exponentials)
    for e in exponentials:
        probabilities.append(e/total)
   
    return probabilities


#Calculates the accuracy of a set of predictions based on the data. finds the label with the highest calculated probability for a training example
#and checks if the example has the same label.
def accuracy(data, predictions):
    correct = 0
    for i in range(len(data)):
        max = -1
        guessed_class = None
        for c in range(len(relevant_genres)):
            if predictions[i][c] > max:
                max = predictions[i][c]
                guessed_class = c
        if data[i]['label'] == guessed_class:
            correct +=1
    return float(correct)/len(data)


#training function for softmax logistic regression. Similar to the training function from HW#2 but the model is a matrix rather than a 1d list.
#prints the training accuracy and validation accuracy for each epoch.
def train(data, valid, epochs, rate, lam):
    valid_accuracies = []
    train_accuracies = []
    models = initialize_models(len(data[0]['features']), len(relevant_genres))
    starting_rate = rate
    for e in range(epochs):
        #rate = starting_rate * math.exp(-.02 * e)
        new_model = []
        probs = [predict(models, d) for d in data]
        for i in range(len(relevant_genres)):
            new_model.append([0] * len(data[0]['features']))
        for c in range(len(relevant_genres)):
            for i in range(len(models[0])):
                new_model[c][i] = models[c][i] - (rate * lam * models[c][i])
                gradient = 0
                count = 0
                for d in data:
                    gradient += (d['features'][i] * ((d['label'] == c) - (probs[count])[c]))
                    count += 1
                new_model[c][i] += (rate * gradient)
        models = new_model
        print("epoch ", e, "complete")
        predictions = [predict(models, p) for p in data]
        #print("Training Accuracy:", accuracy(data, predictions))
        train_accuracies.append(accuracy(data, predictions))
        predictions = [predict(models, p) for p in valid]
        #print("Validation Accuracy:", accuracy(valid, predictions))
        valid_accuracies.append(accuracy(valid, predictions))
    plot.plot(train_accuracies)
    plot.xlabel("Epoch Number")
    plot.ylabel("Training Accuracy")
    plot.title("Training Accuracy vs Epoch Number")
    plot.show()
    plot.plot(valid_accuracies)
    plot.xlabel("Epoch Number")
    plot.ylabel("Validation Accuracy")
    plot.title("Validation Accuracy vs Epoch Number")
    plot.show()    


       
    return models
```
